## 데이터 시각화의 중요성


```python
import seaborn as sns
get_dataset = sns.get_dataset_names()
use_dataset = get_dataset[1]
anscombe = sns.load_dataset(use_dataset)
anscombe
```

```python
anscombe_list = anscombe['dataset'].unique()
anscombe_list
```

```python
import matplotlib.pyplot as plt
plt.plot()
```

```python
d1 = anscombe[anscombe['dataset'] == anscombe_list[0]]
d2 = anscombe[anscombe['dataset'] == anscombe_list[1]]
d3 = anscombe[anscombe['dataset'] == anscombe_list[2]]
d4 = anscombe[anscombe['dataset'] == anscombe_list[3]]
d1,d2,d3,d4
```

```python
plt.plot(d1['x'], d1['y'], 'x')
plt.plot(d2['x'], d2['y'], 'x')
plt.plot(d3['x'], d3['y'], 'x')
plt.plot(d4['x'], d4['y'], 'x')
plt.show()
```

```python
#차트를 한 영역 안에 비교할 수 있도록 모아놓는 기능
fig = plt.figure()

fig.suptitle('Anscombe Dataset으로 보는 데이터 시각화의 의미')
fig.set_tight_layout = True

axes1 = fig.add_subplot(2,2,1) #2row, 2col, position
axes1.plot(d1['x'],d1['y'], 'o')

axes2 = fig.add_subplot(2,2,2)
axes2.plot(d2['x'],d2['y'],'p')

axes3 = fig.add_subplot(2,2,3)
axes3.plot(d3['x'], d3['y'], 'x')

axes4 = fig.add_subplot(2,2,4)
axes4.plot(d4['x'],d4['y'], 'v')
```

## 새로운 데이터 셋
tips : 팁, 청구서 전체 금액, 일행수, 요일, 시간, 흡연 여부, 성별

### matplotlib

```python
use_dataset2 = sns.get_dataset_names()[-2]
tips = sns.load_dataset(use_dataset2)
tips
```

#### 일변량 그래프(변수 하나만 가지고 그리는 그래프)
- 히스토그램  
  구간의 빈도수를 시각화하는 그래프  
  x의 구간을 나누어, y는 빈도수 연속형 변수의 구간별 분포를 나타내는 그래프  
- 막대그래프 : 구간이 연속이 아니고, 이산형임
- 상자그램(boxplot) : 연속형 변수의 분포를 나타내는 그래프(최소 ~ Q1 ~ Q2 ~ Q3 ~ 최대)

```python
tips.columns[0]
```

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.hist(data = tips, x = tips.columns[0], bins = 100) #bin은 막대기 수
axes.set_title('제목')
plt.show()
```

```python
# 막대그래프는 seaborn에서 많이함
sns.countplot(data = tips, x = tips.columns[4])

plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.boxplot(data = tips, x = tips.columns[0])
axes.set_title("제목")
plt.show

#boxplot의 이상치는 중앙의 1.5배를 곱해서 기계적으로 처리해줌
```

#### 이변량 그래프

##### 산점도(scatter plot)
- 2차원 데이터, 두 변수의 상관관계를 표현
- 좌표계의 평면에 데이터를 점으로 표현하는 그래프
- 이상치 확인이 용이함
- 점 하나의 위치는 데이터 하나의 x,y값

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.scatter(x = tips[tips.columns[0]], y = tips[tips.columns[1]]) #총 지불금액과 팁의 산점도
axes.set_title("총 지불금액과 Tip의 상관관계")
plt.show()
```

##### 상자 그림

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.boxplot(
    x = [
            tips[tips['sex'] == 'Female']['tip'],
            tips[tips['sex'] == 'Male']['tip']        
    ],
    labels = ['F', 'M']
   
)
```

#### 다변량 그래프

```python
#map()함수에 사용
colors = {
    'Female' : '#ffffff',
    'Male' : '#000000'    
}
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)

axes1.scatter(
    data = tips,
    x = 'total_bill',
    y = 'tip',
    s = tips['size'] **2*10,
    c = tips['sex'].map(colors),
    alpha = 0.5
)

axes1.set_title('크기와 색깔에 따른 총지출과 tip의 다변량 산점도')
axes1.set_xlabel('총지출')
axes1.set_ylabel('Tip')

fig.suptitle('대제목')

plt.show()
```

```python
#subplots은 튜플로 제공하기 때문에, 언패킹 나머지 하나를 받아줘야함
a, ax = plt.subplots()
sns.countplot(data = tips, x = 'day', palette = 'Accent')
```

#### 이변량 그래프

```python
#subplots은 튜플로 제공하기에, 언패킹 나머지 하나를 받아줘야함
a, ax = plt.subplots()
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', ax = ax, hue = 'sex', palette ='coolwarm')
ax.set_title('지불금액과 tip의 관계')
plt.show
```

```python
a, ax = plt.subplots()
sns.regplot(data = tips, x = 'total_bill', y = 'tip', ax = ax)
ax.set_title('지불금액과 tip의 관계')
plt.show
```

```python
#joint plot
joint = sns.jointplot(data = tips, x ='total_bill', y = 'tip')
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'tip')
joint.figure.suptitle('총 지출과 tips의 joint차트')
plt.show
```

```python
a, ax = plt.subplots()
sns.kdeplot(data = tips, x = 'total_bill', y = 'tip', fill = True, ax = ax)
ax.set_title('KDE plot')
plt.show()
```

```python
#bar plots
a, ax = plt.subplots()
sns.barplot(data = tips, x = 'time', y = 'total_bill', estimator = np.mean, ax = ax)
ax.set_title('Bar plot')
plt.show()

#차트에 보이는 검은색 세로선은 값의 신뢰구간을 표시함
```

```python
a, ax = plt.subplots()
sns.boxplot(data = tips,
            x = 'time',
            y = 'total_bill',
            hue = 'sex',
            hue_order = ['Female','Male'],
            linecolor = '#000000',
            color = 'blue',
            width = 1.0
           )

plt.suptitle('시간대별 지출액')
plt.show()
```

```python
#violin chart
#바이올린 차트는 이상치를 다루지 않음
a, ax = plt.subplots()
sns.violinplot(
    data = tips,
    x = 'time',
    y = 'total_bill',
    hue = 'sex',
    hue_order = ['Female','Male'],
    linecolor = '#ff7777',
    color = 'red',
    width = 1.0   
)

ax.set_title('violin plot')
plt.show()
```

#### pairplot

```python
fig = plt.figure()
sns.pairplot(data = tips)
plt.suptitle('tips 데이터셋의 관계 분석', y = 1.02)
plt.show()
```
