```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.moviechart.co.kr/rank/realtime/index/image'
res = requests.get(url)
res.text
```

```python
movie_soup = BeautifulSoup(res.text, 'html.parser')
movie_soup
```

```python
movie_title = movie_soup.select("div.movie-title > h3 > a")
movie_title_list = [i.text for i in movie_title]
movie_title_list
```

```python
movie_image = movie_soup.select('div.movieBox  a > img')
movie_image_list = [i for i in movie_image]
movie_image_list
```

```python
import re
pattern = re.compile(r'https?://[^\s/$.?#].[^\s]*')
for i in movie_image_list:
    print(pattern.findall(i.__str__()))
```

```python
import os

image_dir = 'images'
if not os.path.exists(image_dir):
    os.makedirs(image_dir)

full_image_dir = os.path.join(os.getcwd(),image_dir)
full_image_dir
```

```python
import re
pattern = re.compile(r'[\\/:"*?<>|]')
```

```python
from PIL import Image
from io import BytesIO

image_list = movie_soup.select('div.movieBox  a  img')
for i in image_list:
    img_response = requests.get("https://www.moviechart.co.kr/"+i['src'])
    print(img_response)
    #print(img_response.content)
    image = Image.open(BytesIO(img_response.content))
    print(image)

    #이미지 이름 지정
    image_name = pattern.sub("",i.text)

    
    image.save(os.path.join(image_dir, i['alt']+'.jpg'))
    
```

```python
for i, j in zip(movie_title_list, image_list):
    img_url = j.get('src')[j.get('src').find('https://'):]
    img_name = pattern.sub('',i)
    image = Image.open(BytesIO(requests.get(img_url).content))
    # 이미지 객체 저장
    image.save(os.path.join(full_image_dir, img_name +".jpg"))
```
