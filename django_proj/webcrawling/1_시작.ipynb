```python
url = 'https://www.google.com/images/branding/googlelogo/1x/googlelogo_color_272x92dp.png'
google_logo = requests.get(url)
google_logo
```

```python
from PIL import Image
from io import BytesIO

Image.open(BytesIO(google_logo.content))
```

```python
import os
image_name = os.path.basename(url)
image_name = image_name[:10] + '.png'
```

```python
# 이미지 저장을 위한 폴더 생성
folder = f'{os.getcwd()}/image'
if not os.path.isdir(folder):
    os.mkdir(folder)
```

```python
#이미지 저장 경로 지정 : folder / image_name
image_path = os.path.join(folder, image_name)
print(image_path)
```

```python
# 이미지 저장
image = open(image_path, 'wb')
image.write(google_logo.content)
image.close()
```
