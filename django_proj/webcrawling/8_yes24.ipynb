```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.yes24.com/Product/Category/BestSeller?categoryNumber=001'
res = requests.get(url)
res
```

```python
yes_soup = BeautifulSoup(res.text, 'html.parser')
yes_soup
```

```python
# 책 제목
# #yesBestList > li:nth-child(1) > div > div.item_info > div.info_row.info_name > a.gd_name
book_title = yes_soup.select('#yesBestList a.gd_name')
book_title_list = [i.text for i in book_title]
book_title_list = [i.text for i in book_title]
```

```python
# 책 저자
# #yesBestList > li:nth-child(1) > div > div.item_info > div.info_row.info_pubGrp > span.authPub.info_auth > a
book_author = yes_soup.select('#yesBestList div.info_row.info_pubGrp span.authPub.info_auth')
book_author_list = [i.text.strip() for i in book_author]
book_author_list
```

```python
# 책 가격(yes24 가격)
# #yesBestList > li:nth-child(24) > div > div.item_info > div.info_row.info_price > strong > em
book_price = yes_soup.select('#yesBestList div.info_row.info_price .yes_b')
book_price_list = [i.text for i in book_price]
book_price_list
```

```python
val_book_list = []
val_author_list = []
val_price_list = []

book_list = {
    'title' : val_book_list,
    'author' : val_author_list,
    'price' : val_price_list,
}
book_list
```

```python
if len(book_title_list) == len(book_author_list) == len(book_price_list):
    for num, i in enumerate(book_title_list):
        for num2, j in enumerate(book_author_list):
            if num == num2:
                for num3, z in enumerate(book_price_list):
                    if num == num3:
                        val_book_list.append(i)                        
                        val_author_list.append(j)                        
                        val_price_list.append(z)
```

```python
import re
pattern = re.compile(r'[\\/:"*?<>|]')
```

```python
import os
image_dir = 'images\yes24'
```

```python
if not os.path.exists(image_dir):
    os.makedirs(image_dir)
```

```python
full_path = os.path.join(os.getcwd(), image_dir)
full_path
```

```python
# 이미지 가져오기
#yesBestList > li:nth-child(1) > div > div.item_img > div.img_canvas > span > span > a > em > img

from PIL import Image
from io import BytesIO

# 이미지 스크래핑
book_cover = yes_soup.select('#yesBestList img')

# 커버랑 이름이랑 매핑함
for i,j in zip(book_cover, book_title_list):

    #URL만 땀
    print(requests.get(i['data-original']).content)
    
    #image 리스트에 담음
    image = Image.open(BytesIO(requests.get(i['data-original']).content))
    print(image)

    # 이름 지정하는데, pattern으로 특수문제 제거함
    image_name = pattern.sub('', j)
    print(image_name)

    #이미지 저장함
    image.save(os.path.join(full_path,j+'.jpg'))
```

```python
class Book:
    def __init__(self, rank, title, author, price):
        self.rank = rank
        self.title = title
        self.author = author
        self.price = price
    
    def __str__(self):
        return f'{self.rank}의 {self.title} - {self.author} - {self.price}'
```

```python
#sqlite DB에 저장
import sqlite3
conn = sqlite3.connect('book3.db')
print(conn)
```

```python
cursor = conn.cursor()
print(cursor)
```

```python
cursor.execute(
    '''
    CREATE TABLE IF NOT EXISTS books(
    Rank INT,
    title TEXT,
    author TEXT,
    price TEXT)
    '''
)
```

```python
book_rank = []
for num, (i,j,z) in enumerate(zip(book_title_list, book_author_list, book_price_list)):
    print(num+1, [i,j,z])
    book_rank.append([num+1, i,j,z])

book_rank
```

```python
for i in book_rank:
    print(i)
    cursor.execute('''
                   INSERT INTO books
                   values (?,?,?,?)
                   '''
                   , i
    )

cursor.execute(
    '''
    select * from books
    '''
)


rows = cursor.fetchall()

for i in rows:
    print(i)
conn.commit()
conn.close()
print('저장완료')
```

```python
import pandas as pd

df_books=pd.DataFrame(book_rank, columns = ['Rank', 'Title', 'Author', 'Price'])

df_books.to_csv('book_list.csv', index = False, encoding = 'utf-8-sig')
```
